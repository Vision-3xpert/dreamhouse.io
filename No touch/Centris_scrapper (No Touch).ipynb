```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
# Initialisez le navigateur
#chrome_options = webdriver.ChromeOptions()
#chrome_options.add_argument("--blink-settings=imagesEnabled=false")
#chrome_options.add_argument("--disable-javascript")
#chrome_options.add_argument("--disable-extensions")
#chrome_options.add_argument("--disable-gpu")
#chrome_options.add_argument("--disable-dev-shm-usage")
#chrome_options.add_argument("--remote-debugging-port=9222")
#driver = webdriver.Chrome(chrome_options=chrome_options)
# Accédez à la page web
#driver.get("https://www.centris.ca/fr/propriete~a-vendre~montreal-ile?view=Thumbnail")
# Attendez que le JavaScript charge la page
#driver.implicitly_wait(10)
# Cliquez sur le bouton
#driver.find_element(by=webdriver.common.by.By.ID, value="ButtonViewSummary").click()
# Récupérez le contenu HTML de la page
#html = driver.page_source
# Utilisez BeautifulSoup pour parser le contenu HTML
#soup = BeautifulSoup(html, "html.parser")
# Rechercher le nombre de propriété sur lesquels itéré.


titre = []
price = []
address = []
sqfeet_building = []
sqfeet_terrain = []
piece  = [] 
car_score = []
cafe_score = []
park_score = []
walk_score = []
quiet_score = []
cycling_score = []
primary_score = []
transit_score = []
vibrant_score = []
daycares_score = []
groceries_score = []
nightlife_score = []
shopping_score = []
restaurant_score = []
high_school_score = []
quartiers = ["Montreal-Est", "Montreal-Ouest", "Montreal-Ahuntsic-Cartierville"] 
quartiers_1 = ["Montreal-Anjou", "Montreal-Cote-des-neiges-Notre-Dame-de-Grace",
"Montreal-L-Ile-Bizard-Sainte-Genevieve"]
quartiers1 = ["Montreal-LaSalle", "Montreal-Lachine", "Montreal-Le-Plateau-Mont-Royal", "Montreal-Le-Sud-Ouest"]
quartiers1_1 = ["Montreal-Mercier-Hochelaga-Maisonneuve", "Montreal-Montreal-Nord", "Montreal-Outremont", "Montreal-Pierrefonds-Roxboro"]
quartiers2= ["Montreal-Riviere-des-Prairies-Pointe-aux-Trembles", "Montreal-Rosemont-La-Petite-Patrie"]
quartiers2_1 = ["Montreal-Saint-Laurent", "Montreal-Saint-Leonard", "Montreal-Verdun-Ile-des-Soeurs"]
quartiers3 = ["Montreal-Ville-Marie"]
quartiers3_1 = ["Montreal-Villeray-Saint-Michel-Parc-Extension", "Westmount" "Kirkland"]
quartiers4 = ["baie-d-urfe", "Pointe-Claire" ,"Cote-Saint-Luc"]
def get_property_from_quartier(quartier):
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument("--blink-settings=imagesEnabled=false")
    driver = webdriver.Chrome(chrome_options=chrome_options)
    # Accédez à la page web
    driver.get(f"https://www.centris.ca/fr/propriete~a-vendre~{quartier}?view=Thumbnail")
    # Attendez que le JavaScript charge la page
    driver.implicitly_wait(10)
    # Cliquez sur le bouton
    driver.find_element(by=webdriver.common.by.By.ID, value="ButtonViewSummary").click()
    time.sleep(5)
    # Récupérez le contenu HTML de la page
    html = driver.page_source
    # Utilisez BeautifulSoup pour parser le contenu HTML
    soup = BeautifulSoup(html, "html.parser")
    # Rechercher le nombre de propriété sur lesquels itéré.
    resultCount = soup.find("span", {"class": "resultCount"}).text
    resultCount = int(resultCount.replace(" ",""))

    for i in range(resultCount):
        Titre = soup.find("span", {"data-id": "PageTitle"}).text
        titre.append(Titre)
        Price = soup.find("span", {"id": "BuyPrice"}).text
        price.append(Price)
        Address = soup.find("h2", {"itemprop":"address"}).text
        address.append(Address)

        car_friendly = soup.find("li",{"data-id":"car_friendly"})
        if car_friendly is None:
            car_score.append("N/A")
        else:
            car_score.append(car_friendly.find("span",{"class":"ll-score__badge"}).text)

        primary_schools = soup.find("li",{"data-id":"primary_schools"})
        if primary_schools is None:
            primary_score.append("N/A")
        else:
            primary_score.append(primary_schools.find("span",{"class":"ll-score__badge"}).text)
            
        quiet = soup.find("li",{"data-id":"quiet"})
        if quiet is None:
            quiet_score.append("N/A")
        else:
            quiet_score.append(quiet.find("span",{"class":"ll-score__badge"}).text)
            
        parks = soup.find("li",{"data-id":"parks"})
        if parks is None:
            park_score.append("N/A")
        else:
            park_score.append(parks.find("span",{"class":"ll-score__badge"}).text)
            
        groceries = soup.find("li",{"data-id":"groceries"})
        if groceries is None:
            groceries_score.append("N/A")
        else:
            groceries_score.append(groceries.find("span",{"class":"ll-score__badge"}).text)
            
        transit_friendly = soup.find("li",{"data-id":"transit_friendly"})
        if transit_friendly is None:
            transit_score.append("N/A")
        else:
            transit_score.append(transit_friendly.find("span",{"class":"ll-score__badge"}).text)
            
        pedestrian_friendly = soup.find("li",{"data-id":"pedestrian_friendly"})
        if pedestrian_friendly is None:
            walk_score.append("N/A")
        else:
            walk_score.append(pedestrian_friendly.find("span",{"class":"ll-score__badge"}).text)
            
        restaurants = soup.find("li",{"data-id":"restaurants"})
        if restaurants is None:
            restaurant_score.append("N/A")
        else:
            restaurant_score.append(restaurants.find("span",{"class":"ll-score__badge"}).text)
            
        shopping = soup.find("li",{"data-id":"shopping"})
        if shopping is None:
            shopping_score.append("N/A")
        else:
            shopping_score.append(shopping.find("span",{"class":"ll-score__badge"}).text)
            
        cycling_friendly = soup.find("li",{"data-id":"cycling_friendly"})
        if cycling_friendly is None:
            cycling_score.append("N/A")
        else:
            cycling_score.append(cycling_friendly.find("span",{"class":"ll-score__badge"}).text)
            
            
        vibrant = soup.find("li",{"data-id":"vibrant"})
        if vibrant is None:
            vibrant_score.append("N/A")
        else:
            vibrant_score.append(vibrant.find("span",{"class":"ll-score__badge"}).text)
            
        cafes = soup.find("li",{"data-id":"cafes"})
        if cafes is None:
            cafe_score.append("N/A")
        else:
            cafe_score.append(cafes.find("span",{"class":"ll-score__badge"}).text)
            
        high_schools = soup.find("li",{"data-id":"high_schools"})
        if high_schools is None:
            high_school_score.append("N/A")
        else:
            high_school_score.append(high_schools.find("span",{"class":"ll-score__badge"}).text)
            
        nightlife = soup.find("li",{"data-id":"nightlife"})
        if nightlife is None:
            nightlife_score.append("N/A")
        else:
            nightlife_score.append(nightlife.find("span",{"class":"ll-score__badge"}).text)
            
        daycares = soup.find("li",{"data-id":"daycares"})     
        if daycares is None:
            daycares_score.append("N/A")
        else:
            daycares_score.append(daycares.find("span",{"class":"ll-score__badge"}).text)

        Carac_val = soup.find_all("div", {"class":"carac-title"})
        superficie_building = None
        superficie_terrain = None
        for c in Carac_val:
            if "Superficie brute" in c.text or "Superficie nette" in c.text or "Superficie du bâtiment (au sol)" in c.text:
                superficie_building = (c.find_next_sibling("div").text)
            if "Superficie du terrain" in c.text:
                superficie_terrain  = (c.find_next_sibling("div").text)
        if superficie_building is None:
            sqfeet_building.append("N/A")
        else:
            sqfeet_building.append(superficie_building)
        if superficie_terrain is None:
            sqfeet_terrain.append("N/A")
        else:
            sqfeet_terrain.append(superficie_terrain)
    #        if c.text == "Superficie du terrain":
    #            sqfeet_terrain.append(c.find_next_sibling("div").text)
    #        elif c.text !="Superficie du terrain":
    #            sqfeet_terrain.append("N/A")


        Piece = soup.select_one(".piece")
        if Piece is not None:
            piece.append(Piece.get_text(strip=True))
        else:
            piece.append("N/A")

        #    driver.delete_all_cookies()
        driver.find_element(by=webdriver.common.by.By.CSS_SELECTOR ,value=".next").click()
        time.sleep(0.5)

        # Récupérez le contenu HTML de la page mise à jour
        html = driver.page_source
        soup = BeautifulSoup(html, "html.parser")
        #price = price.replace("\xa0"," ")
```

```python
import pandas as pd
Addresses = pd.DataFrame(address, columns=["Addresse"])
Type_propriété = pd.DataFrame(titre, columns=["titre"])
Prix_Vente = pd.DataFrame(price, columns=["price"])
Nb_Pieces = pd.DataFrame(piece, columns=["piece"])
PC_Appart = pd.DataFrame(sqfeet_building, columns=["sqfeet_buildinge"])
PC_Terrain = pd.DataFrame(sqfeet_terrain, columns=["sqfeet_terraine"])
car_point = pd.DataFrame(car_score, columns=["car_scoree"])
cafe_point = pd.DataFrame(cafe_score, columns=["cafe_scoree"])
park_point = pd.DataFrame(park_score, columns=["park_scoree"])
walk_point = pd.DataFrame(walk_score, columns=["walk_scoree"])
quiet_point = pd.DataFrame(quiet_score, columns=["quiet_scoree"])
cycling_point = pd.DataFrame(cycling_score, columns=["cycling_scoree"])
primary_point = pd.DataFrame(primary_score, columns=["primary_scoree"])
transit_point = pd.DataFrame(transit_score, columns=["transit_scoree"])
vibrant_point = pd.DataFrame(vibrant_score, columns=["vibrant_scoree"])
daycares_point = pd.DataFrame(daycares_score, columns=["daycares_scoree"])
groceries_point = pd.DataFrame(groceries_score, columns=["groceries_scoree"])
nightlife_point = pd.DataFrame(nightlife_score, columns=["nightlife_scoree"])
shopping_point = pd.DataFrame(shopping_score, columns=["shopping_scoree"])
restaurant_point = pd.DataFrame(restaurant_score, columns=["restaurant_scoree"])
high_school_point = pd.DataFrame(high_school_score, columns=["high_school_scoree"])
Centris_data = pd.concat([Addresses,Type_propriété,Prix_Vente,Nb_Pieces,PC_Appart,PC_Terrain,car_point,cafe_point,walk_point,park_point], axis=1)
Centris_data2 = pd.concat([quiet_point,cycling_point,transit_point,vibrant_point,daycares_point,primary_point,shopping_point,groceries_point,nightlife_point,restaurant_point], axis=1)
Centris_data3 = pd.concat([Centris_data,Centris_data2,high_school_point], axis=1)
Centris_data3
```
