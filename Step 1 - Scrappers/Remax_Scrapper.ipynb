```python
# Import de la bibliothèque Selenium
from urllib import response
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
from bs4 import BeautifulSoup

# Création d'un objet webdriver pour Chrome
driver = webdriver.Chrome()

# Navigation vers la page web
driver.get("https://www.remax-quebec.com/fr/a-vendre/montreal")

# Attente de 10 secondes pour laisser le temps au JavaScript de charger la page
driver.implicitly_wait(5)

# Récupération du contenu HTML de la page
html = driver.page_source

# Parsing du contenu HTML avec BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')

# Recherche de l'élément à faire défiler
Price = driver.find_element(by=webdriver.common.by.By.CSS_SELECTOR, value=".card")

# Boucle while qui s'exécute tant que l'élément est visible
start_time = time.time()

while Price.is_displayed():
    # Envoi de la touche "Page Down" au navigateur
    elapsed_time = time.time() - start_time #

    Price.send_keys(Keys.PAGE_DOWN)
    if elapsed_time > 150:
        break

# Exécution de la fonction JavaScript qui défile l'élément jusqu'en bas
price_house = driver.find_elements(by=webdriver.common.by.By.CSS_SELECTOR, value=".property-details--price")
Type_Logement = driver.find_elements(by=webdriver.common.by.By.CSS_SELECTOR, value=".property-details--title")
Adresse = driver.find_elements(by=webdriver.common.by.By.CSS_SELECTOR, value=".property-details--address")
icons = soup.find_all('div', class_='property-details--icons')
Bedrooms_list = []
Toilet_list= []
Bathrooms_list = []
SqFoot_List = []
for icon in icons:
    if icon.find('span',attrs={"aria-label":"bed icon"}) is not None:
        Bedrooms_list.append(icon.find('span',attrs={"aria-label":"bed icon"}))
    else:
        Bedrooms_list.append("N/A")

    if icon.find('span',attrs={"aria-label":"bath icon"}) is not None:
        Bathrooms_list.append(icon.find('span',attrs={"aria-label":"bath icon"}))
    else:
        Bathrooms_list.append("N/A")
    if icon.find('span',attrs={"aria-label":"door-open icon"}) is not None:
        SqFoot_List.append(icon.find('span',attrs={"aria-label":"door-open icon"}))
    else:
        SqFoot_List.append("N/A")
    if icon.find('span', attrs={'aria-label:"toilet icon'}) is not None:
        Toilet_list.append(icon.find('span', attrs={'aria-label:"toilet icon'}))
    else: 
        Toilet_list.append("N/A")
Adresses = []
for Adress_unit in Adresse:
    Adresses.append(Adress_unit.text)
# Affichage du prix
Type_Log = []
for Logement_unit in Type_Logement:
    Type_Log.append(Logement_unit.text)
prix = []
for price_unit in price_house:
    prix.append(price_unit.text)
#print(len(prix),Type_Log,Adresses)

# Fermeture du navigateur
driver.close()
```

```python
import re
nb_bed = []
for x in Bedrooms_list:
    if x !="N/A":
        z=(x.find_next_sibling('label',class_='medium medium--normal').text)
    else:
        z=x
    nb_bed.append(z.strip())
nb_bath = []
for x in Bathrooms_list:
    if x !="N/A":
        z=(x.find_next_sibling('label',class_='medium medium--normal').text)
    else:
        z=x
    nb_bath.append(z.strip())
nb_SqFoot = []
for x in SqFoot_List:
    if x !="N/A":
        z=(x.find_next_sibling('label',class_='medium medium--normal').text)
    else:
        z=x
    nb_SqFoot.append(z.strip())
nb_Toilet = []
for x in Toilet_list:
    if x != "N/A":
        z = (x.find_next_sibling('label',class_='medium medium--normal').text)
    else:
        z=x
    nb_Toilet.append(z.strip())
```

```python
## Nettoyage des données

## Nettoyage du nombre de chambres 
bedrooms_propre = []
for item in nb_bed:
    bed_propre = re.sub(r'\(.*\)', '', item)
    bedrooms_propre.append(bed_propre)
print(bedrooms_propre)

##Nettoyage du square footage 
SQUARE_METER_TO_SQUARE_FOOT = 10.764
sqfeet_propre = []

for value in nb_SqFoot:
# Recherche de tous les espaces entre les chiffres
    matches = re.finditer(r'\d{1,3}(?:[., ]\d{3})*[.,]?\d+', value)

    for match in matches:
    # Remplacement des virgules par des points
        decimal = match.group().replace(",", ".")
        decimal = decimal.replace(".", ".")
        decimal = decimal.replace(" ","")
        value = value.replace(match.group(), decimal)

# Remplacement de tous les espaces dans la chaîne
    value = re.sub(r'(?<=\d)(\s)(?=\d)', "", value)


    # Remplacement de la partie numérique dans la chaîne originale
    test = value.split()
    if test[-1] == "m²":
        test[0] = int(round(float(test[0])*SQUARE_METER_TO_SQUARE_FOOT,0))
        test[-1] = "pi²"
    if test[-1] == "pi²":
        test[0] = int(round(float(test[0]),0))
    sqfeet_propre.append(test[0])
# Affiche "1263.98 mètres et 24.567 mètres"
print(sqfeet_propre)


#  Nettoyage du prix 
cleaned_prices = []
for price in prix:
    if "TPS/TVQ" in price:
        price = price.replace("$ +TPS/TVQ", "")
        price = int(price.replace(" ", "").replace("$", ""))
        price = price*1.15
    elif "ou" in price:
        price = re.sub(r' ou.*', "", price)
        price= int(price.replace(" ","").replace("$",""))
    else:
        price = int(price.replace(" ", "").replace("$", ""))
    cleaned_prices.append(int(price))

print(cleaned_prices)
```
